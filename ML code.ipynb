```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df=match_df[match_df["dl_applied"]==0]
```

```python
match_df=match_df[["match_id",'city','winner','total_runs']]
```

```python
delivery_df=match_df.merge(delivery,on="match_id")
```

```python
delivery_df=delivery_df[delivery_df["inning"]==2]
```

```python
delivery_df.shape
```

```python
delivery_df['current_score']=delivery_df.groupby("match_id")["total_runs_y"].cumsum()
```

```python
delivery_df["runs_left"]=delivery_df["total_runs_x"]-delivery_df["current_score"]
```

```python
delivery_df["balls_left"]=126-(delivery_df['over']*6+delivery_df["ball"])
```

```python
delivery_df
```

```python
# 1. Replace NaN with the *integer* 0
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna(0)

# 2. Wicket flag: anything that is *not equal to 0* is a dismissal
delivery_df['is_wicket'] = (delivery_df['player_dismissed'] != 0).astype(int)

# 3–4. Same as above
delivery_df = delivery_df.sort_values(['match_id', 'inning', 'over', 'ball'])
delivery_df['wickets_lost'] = (
    delivery_df
      .groupby(['match_id', 'inning'])['is_wicket']
      .cumsum()
)
delivery_df['wickets_left'] = 10 - delivery_df['wickets_lost']
delivery_df.head()
```

```python
# -------------------------------------------------------------
# 0)  Load (or re-load) the file so nothing is pre-modified
# -------------------------------------------------------------
# delivery_df = pd.read_csv("/mnt/data/4634ea54-340e-4098-a60a-71f7031ebf70.csv")

# -------------------------------------------------------------
# 1)  Flag genuine wickets
#     • we use **dismissal_kind**, so “retired hurt” is ignored
# -------------------------------------------------------------
wicket_kinds = [
    "bowled", "caught", "caught and bowled", "lbw",
    "run out", "stumped", "hit wicket",
    "obstructing the field", "handled the ball", "hit the ball twice"
]

delivery_df["is_wicket"] = delivery_df["dismissal_kind"].isin(wicket_kinds).astype(int)

# -------------------------------------------------------------
# 2)  Keep only normal innings (1 & 2).  Super-overs are 3-4-5.
# -------------------------------------------------------------
delivery_df = delivery_df[delivery_df["inning"].isin([1, 2])].copy()

# -------------------------------------------------------------
# 3)  Sort in true ball order  ➜  match ▸ inning ▸ over ▸ ball
# -------------------------------------------------------------
delivery_df = (
    delivery_df
      .sort_values(["match_id", "inning", "over", "ball"])
      .reset_index(drop=True)
)

# -------------------------------------------------------------
# 4)  Running total of wickets lost inside EACH innings
# -------------------------------------------------------------
delivery_df["wickets_lost"] = (
    delivery_df
      .groupby(["match_id", "inning"])["is_wicket"]
      .cumsum()
)

# -------------------------------------------------------------
# 5)  Wickets still in hand  (starts at 10, never below 0)
# -------------------------------------------------------------
delivery_df["wickets_left"] = 10 - delivery_df["wickets_lost"]
delivery_df
```

```python
delivery_df['crr']=delivery_df['current_score']*6/(120-delivery_df['balls_left'])
```

```python
delivery_df['rrr']=(delivery_df['runs_left']*6)/delivery_df["balls_left"]
```

```python
delivery_df
```

```python
def result(row):
    return 1 if row["batting_team"]==row["winner"] else 0
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
```

```python
final_df=delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','total_runs_x','crr','rrr','result']]
```

```python
final_df=final_df.sample(final_df.shape[0])   # To shuffle it so that the model doesn't Create any bias
```
